```python
# Function to plot Pareto chart for a given index group
def plot_pareto(df, index_value):
    # Filter the data for the given index group
    group_data = df.loc[index_value]

    # Sum the values for each feature/column
    feature_sums = group_data.sum()

    # Sort values in descending order for Pareto
    sorted_features = feature_sums.sort_values(ascending=False)

    # Cumulative sum for Pareto line plot
    cumulative_sum = sorted_features.cumsum()
    cumulative_percentage = 100 * cumulative_sum / cumulative_sum.iloc[-1]

    # Plotting
    fig, ax1 = plt.subplots(figsize=(10, 6))

    # Bar plot
    sorted_features.plot(kind='bar', color='skyblue', ax=ax1)
    ax1.set_ylabel('Sum of Feature Values')
    ax1.set_title(f'Pareto Chart for Region_Channel {index_value}')

    # Line plot on right Y-axis
    ax2 = ax1.twinx()
    cumulative_percentage.plot(color='red', marker='o', ax=ax2)
    ax2.set_ylabel('Cumulative Percentage (%)')

    plt.show()


plot_pareto(customer_data_region_channel, '3_1')  
plot_pareto(customer_data_region_channel, '3_2')  
plot_pareto(customer_data_region_channel, '2_1')  
plot_pareto(customer_data_region_channel, '2_2')  
plot_pareto(customer_data_region_channel, '1_1')  
plot_pareto(customer_data_region_channel, '1_2')  
```

One problem with the dataset is the value ranges are remarkably different across various categories (e.g. `Fresh` and `Grocery` compared to `Detergents_Paper` and `Delicassen`). Diverse value ranges in different features could cause issues in our clustering. The way to reduce the problem is through feature scaling. We'll use this technique again with this dataset.

#### We will use the `StandardScaler` from `sklearn.preprocessing` and scale our data. Read more about `StandardScaler` [here](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler).

*After scaling your data, assign the transformed data to a new variable `customers_scale`.*

```python
from sklearn.preprocessing import StandardScaler
features = customer_data_region_channel_filtered.select_dtypes(include=[np.number]).columns
scaler = StandardScaler()
customer_scale = customer_data_region_channel_filtered.copy()
customer_scale[features] = scaler.fit_transform(customer_data_region_channel_filtered[features])
print(customer_scale[features].mean())  # Should be close to 0
print(customer_scale[features].std())   # Should be close to 1
```

# Challenge 4 - Data Clustering with K-Means

Now let's cluster the data with K-Means first. Initiate the K-Means model, then fit your scaled data. In the data returned from the `.fit` method, there is an attribute called `labels_` which is the cluster number assigned to each data record. What you can do is to assign these labels back to `customers` in a new column called `customers['labels']`. Then you'll see the cluster results of the original data.

```python
# Using Elbow Method to find optimal 'k'
from sklearn.cluster import KMeans

inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(customer_scale)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia)
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

# Challenge 5 - Data Clustering with DBSCAN

Now let's cluster the data using DBSCAN. Use `DBSCAN(eps=0.5)` to initiate the model, then fit your scaled data. In the data returned from the `.fit` method, assign the `labels_` back to `customers['labels_DBSCAN']`. Now your original data have two labels, one from K-Means and the other from DBSCAN.

```python
from sklearn.cluster import DBSCAN 

dbscan = DBSCAN(eps=1.9000000000000002, min_samples=4)  # eps and min_samples are hyperparameters to be tuned
labels = dbscan.fit_predict(customer_scale)
customer_scale['DBSCAN_Labels'] = labels
labels_DBSCAN = customer_scale['DBSCAN_Labels'].value_counts()

# Display the count of instances in each cluster
print(labels_DBSCAN)
```

```python

df = customer_scale.copy()
# Assuming df is your DataFrame with numeric features prepared
# Remove 'DBSCAN_Labels' if already present to avoid confusion
df_cleaned = df.drop(columns=['DBSCAN_Labels'], errors='ignore')

# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_cleaned[['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']])

# Try different values of `eps` and `min_samples`
best_eps = None
best_min_samples = None
least_noise_ratio = np.inf

# Analyze over a range of eps and min_samples values
for eps in np.arange(0.2, 1.0, 0.1):
    for min_samples in range(2, 10):
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(X_scaled)
        
        # Calculate noise ratio
        noise_ratio = list(labels).count(-1) / len(labels)
        print(f'eps={eps}, min_samples={min_samples}, Noise ratio: {noise_ratio:.2f}')

        # Save the best configuration
        if noise_ratio < least_noise_ratio:
            least_noise_ratio = noise_ratio
            best_eps = eps
            best_min_samples = min_samples

print(f'Optimal eps: {best_eps}, min_samples: {best_min_samples}')

# Apply DBSCAN with optimal parameters
dbscan_optimal = DBSCAN(eps=best_eps, min_samples=best_min_samples)
optimal_labels = dbscan_optimal.fit_predict(X_scaled)

# Add the resulting labels to the DataFrame
df_cleaned['DBSCAN_Labels'] = optimal_labels

# Display the adjusted DataFrame
print(df_cleaned)
```
